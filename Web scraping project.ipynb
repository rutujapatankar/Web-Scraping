# Web Scraping of Best 30 Grocery Stores in New York

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
url = "https://www.yellowpages.com/new-york-ny/grocery-stores"

page = requests.get(url)

soup = BeautifulSoup(page.content,'html.parser')
```

```python
Shop_name_tags = soup.select(".business-name")

Shop_name = []
for i in Shop_name_tags:
    Shop_name.append(i.get_text())
print(Shop_name)
```

```python
len(Shop_name)
```

```python
print(Shop_name[:30])
```

```python
len(Shop_name[:30])
```

```python
Shop_types = soup.select(".categories")

Shop_type = []
for i in Shop_types:
    Shop_type.append(i.get_text())
print(Shop_type)
```

```python
len(Shop_type)
```

```python
print(Shop_type[:30])
```

```python
len(Shop_type[:30])
```

```python
address = soup.select(".street-address")

Street_address = []
for i in address:
    Street_address.append(i.get_text())
print(Street_address)
```

```python
len(Street_address)
```

```python
locality = soup.select(".locality")

Locality_add = []
for i in locality:
    Locality_add.append(i.get_text())
print(Locality_add)
```

```python
len(Locality_add)
```

```python
phone = soup.select(".phones.phone.primary")

Phone_number= []
for i in phone:
    Phone_number.append(i.get_text())
print(Phone_number)
```

```python
len(Phone_number)
```

```python
df = pd.DataFrame({"Shop Name":Shop_name[:30],"Shop Type":Shop_type[:30],"Street Address":Street_address,"Locality":Locality_add,"Contact No":Phone_number})
df
```

```python
df.to_csv("Top 30 Grocery Stores.csv",index = False)
```
